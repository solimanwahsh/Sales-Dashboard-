```python
ga=df[["date","branch","cogs"]]
gan=ga.set_index("date")
gan=ga.groupby(["date","branch"],as_index=False)["cogs"].sum()
gan=gan.pivot(index="date",columns="branch",values="cogs") ###also foucs in this 
gan
gan.plot(figsize=(22,6)) ##foucs here we droped in there 
sns.lineplot(data=gan)
```

```python
norm=gan/gan.iloc[0,:]  ## we have to do normalization to start with same to know any branch is prograss to another ,here we devied all rows on first row 

```
