When calculating clay and shale volumes we often use the percentiles as our interpretation parameters.

These can be calculated using built in pandas functions: mean() and quantile().

```python

mean = df['GR'].mean()
p5 = df['GR'].quantile(0.05)
p95 = df['GR'].quantile(0.95)            # Notice we utilize Quantile function to limit a range which need it by percentage  

print(f'Mean: \t {mean}')
print(f'P05: \t {p5}')
print(f'P95: \t {p95}')
```

To get an idea of where these points fall in relation to our data, we can add them onto the plot using axvline 
and passing in the calculated variables, a colour and a label.





```python
plt.figure(figsize=(12,8))
df["GR"].plot(kind="hist",bins=30,color="r",alpha=0.6,edgecolor="black",density=True)
df["GR"].plot(kind="kde",color="black")
plt.xlabel("Gamma Ray",fontsize=14)
plt.ylabel("Frequency",fontsize=14)

plt.axvline(mean,color="green",label="mean")
plt.axvline(p5,color="blue",label="P5 percentile")
plt.axvline(p95,color="orange",label="P95 percentile")    

plt.legend()
plt.xlim(0,200)
```

```python
plt.figure(figsize=(5,8))
plt.plot(df["GR"],df["DEPTH"],c="black",lw=0.5)
plt.fill_betweenx(df["DEPTH"],50,df["GR"],where=df["GR"]<=50,facecolor="green" )
plt.fill_betweenx(df["DEPTH"],df["GR"],50,where=df["GR"]>=50 ,facecolor="yellow" )


plt.xlim(0,150)
plt.ylim(4400, 4000)
```

```python
left_value=0
right_value=150
span=abs(left_value-right_value)   ## don't forget absolutly 

cmap=plt.get_cmap('hot_r')

color_index=np.arange(left_value,right_value,span/100)   # span\100 that means 1.5
```

```python
color_index
```

```python
plt.figure(figsize=(5,8))
plt.plot(df["GR"],df["DEPTH"],c="black",lw=0.5)

plt.xlim(0,150)
plt.ylim(4400, 4000)

for index in (color_index):
    index_value = (index - left_value)/span
    color = cmap(index_value)
#     print(f'{index}\t{index_value}\t{color}')
    plt.fill_betweenx(df['DEPTH'], left_value, df['GR'], where=df['GR']>=index, color=color)
```

```python
plt.figure(figsize=(5,8))
plt.plot(df["GR"],df["DEPTH"],c="black",lw=0.5)

plt.xlim(0,150)
plt.ylim(4400, 4000)

for index in sorted(color_index):
    index_value = index/span
    color = cmap(index_value)
#     print(f'{index}\t{index_value}\t{color}')
    plt.fill_betweenx(df['DEPTH'], right_value, df['GR'], where=df['GR']>=index,color=color)
    
```
