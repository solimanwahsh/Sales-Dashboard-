```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
#import plotly.express as px

df=pd.read_csv("C:/Users/FRIENDS/Desktop/.vs/Madriad cleaning.csv")

df.info()

df.head(18)

plt.figure(figsize=(18,8))
for e,i in enumerate(df.columns[2:]):
    plt.subplot(2,3,e+1)
    plt.title("Distribution of "+ i)
    sns.histplot(df[i],kde=True)
    plt.tight_layout()
```

```python
plt.figure(figsize=(18,8))
for e,i in enumerate (df.columns[2:]):
    plt.subplot(2,3, e+1)
    plt.title("Distribution of "+ i)
    sns.boxplot(df[i])
    plt.tight_layout()
```

```python
plt.bar(df.distric,color="r",height=1)
```

```python
corr=df[["sq_mt_built","n_rooms","n_bathrooms"]]
```

```python
sns.pairplot(corr,kind='reg')

plt.tight_layout()

```

```python
sns.heatmap(corr.corr(),annot=True)
plt.title("correlation coeffecient between numerical columns")
plt.tight_layout()
```

```python
sns.regplot(data=df,x="sq_mt_built",y="buy_price")
plt.title("correlation between sq_mt_built and buy price")
plt.xlabel("sq_mt_built")
plt.ylabel("buy price")
```

```python
cor=df[["sq_mt_built","buy_price","n_rooms","n_bathrooms"]]
sns.heatmap(cor.corr(),annot=True)
```

```python
df=df[['id', 'district', 'sq_mt_built', 'n_rooms', 'n_bathrooms', 'buy_price', 'buy_price_by_area']]
```

```python
df.rename(columns={"distric":"district"},inplace=True)
```

```python
df=df[['id', 'district', 'sq_mt_built', 'n_rooms', 'n_bathrooms', 'buy_price', 'buy_price_by_area']]
```

```python
df.head()
```

```python
##t.figure(figsize=(12,10))
meanprice=df.groupby("district")["buy_price"].mean().round(3).sort_values(ascending=False)
meanprice.head(10).plot(kind="bar",xlabel="district",ylabel="buy_price",color="b",grid=True)
#lt.tight_layout()
```

```python
##t.figure(figsize=(12,10))
meanprice=df.groupby("district")["buy_price_by_area"].mean().round(1).sort_values(ascending=False)
meanprice.head(10).plot(kind="bar",xlabel="district",ylabel="buy_price_by_area",color="b",grid=True)
#lt.tight_layout()
```
